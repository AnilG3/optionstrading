## Working with Intraday Price Spike Charts

### Key Concepts

Measure volatility of a series of price changes over different time frames, and visualize as histogram. Histogram are useful in identifying different type of price change behavior

* Rising instability that preceds a large change,
* Decreasing volatility near the end of a sharp up/down trend,
* Sudden up/down or down/up transitions preceding change in direction.

Frequency and magnitude of large price changes in smaller time frames is much larger than predicted by Option pricing theory.

### Introduction

* Volatility often reveals more about the behavior of a security than any other technical parameter.
    - Volatility rises when prices become unstable; when prices become unstable they usually fall. *CBOE Volatility Index (VIX)*
    - Study price changes in context of volatility, i.e. measure in standard deviation.
    - An *orderly market* can be built on a series of consistent, but large, price changes, whereas a *disorderly market* is often characterized by a series of small inconsistent changes.
    - Rising level of instability often marks the end of a brief but sharp uptrend. This can be visualized as increasingly large up and down price spikes. Sometimes, single pair of spikes in opposite directions.
    - A sharp consistently downward trend often ends with successively smaller downward price spikes followed by a significant upward spike.


* Principal goal of option trader:
    - Arbitrage subtle differences between implied and fair volatility by structuring positions that capitalize on these discrepancies.
    - Buying and selling standard deviations.

### Objective

* A charting strategy that facilitates this type of analysis.
* Recasting minute-by-minute price changes in standard deviations and displaying the results as a histogram at the bottom of a stock chart.

*Goal is to spot differential volatility where up and down spikes occur with distinctly different frequency, and reliably predict the end of an up or down trend.*

### Basic Calculations for Annual Volatility Estimate

1. Estimate annual volatility by calculating the standard deviation of a series of price changes,
2. Multiplying the result from `1` by an annualization factor equal to square root of the the number of individual price changes in a single year.
3. Calculation is based on the log of the price change

Calculation of Annual Volatility Estimate can be considered consist of two parts:

1. Determine value of a 1-day 1 standard deviation changes (dollar value), and
2. the annualization of result in `1`.

The log of each price change is used because of lognormal price distribution, skewed price distribution so that no price ever falls below zero. Stated differently, the logs of the price changes are distributed according to normal distribution curve.

#### Calculate Price Spike in Standard Deviations

Using a rolling window, we can continuously calculate the value of 1 day 1 StdDev change and display those values as a histogram at the bottom of a security chart.

### Spikes and Volatility using Options
The dynamics change significantly when the time between closings is reduced from 1 day to 1 minute. Relationship between value of 1-day 1-StdDev change and annual volatility does not persist at single minute level. In the very short time frame of a few minutes, options can dramatically underprice risk.

The distortion can be visualized measuring short term price changes against the implied volatility of a stock. Price spikes can be calculated in standard deviations using implied volatility for at-the-money options as a reference instead of sliding window of price changes. Thus allows us to compare theoretical predictions derived from option prices with actual realized volatility.

### 2 Minutes Period Stock Data

#### Intraday Predicted Spikes using 2 minutes Price Data and ATM Implied Volatility

For each 2-minute interval, the value of a 1 StdDev change was calculated using fixed implied volatility of ATM strike and the current closing price. The difference between realized and predicted frequency is very large across the entire group.

### Price Spikes at the Single Minute Level

Because annual volatility and individual price changes become decoupled in very brief time frames, it is more reasonable to measure each price change against the most recent. For example, if price changes are very modest for an extended time period, standard deviation value falls, increasing senstivity to rising short term volatility. This measurement is important for detecting the top of an uptrend after the market has experienced a steady orderly series of gradual increases. Conversely, if the reference is fixed using annual volatility, subtle increases will be difficult to detect.

Customize the analysis by slightly adjusting the use of sliding window. The most reasonable modification delays moving the window forward after every second calculation so that each pair of price changes can be measured using the same reference value for a 1 StdDev change. It solves a problem that is caused when one large spike diminishes the value of the next by increasing the size of 1 StdDev refernece at the end of window.

```python
##### Functions - Stock #####

# Calculated Price Spikes in Standard Deviation
def calc_price_spikes_annual_vol(df, window=20, yearly_trading_period=252):
    # ----------------------------------------------------------------
    # Calculate Daily Price Spikes and Annual Volatility Close to Close (C-C)
    # ----------------------------------------------------------------
    # df = daily or minute price data
    # window = sliding window to use for calculating standard deviation
    # yearly_trading_period = Number of trading periods in a year
    #   For daily pricing data, it is number of trading days in a year = 252
    #   For 2 minute pricing data, it is number of 2 minutes trading intervals in a year
    #   2 minute interval per year = 252 days x 24 hours x 30 per hour = 181,440

    # Reset index before calculations otherwise warnings are issued
    index_name = df.index.name
    df = df.reset_index()
    
    # Calculate Percent Change from previous day
    df['PctChng'] = ((df.Close / df.Close.shift(1)) - 1.0) * 100.0
    
    # Calculate Log Change from previous day
    # Table 4.1, Column B
    df['LogChng'] = np.log(df.Close / df.Close.shift(1))
    
    # Standard Deviation of LogChng over a specified time period (sliding window)
    # Table 4.1 Column C
    # The value calculated is equal to standard deviation of the logs 
    # of the X most recent price changes (Column B) multiplied by the closing
    # price on current day. It represents, in dollar and cents, the value of 
    # a 1-day 1 StdDev price change.
    df['StdDevLog'] = df.LogChng.rolling(window).std() * df.Close
    
    # Price Spikes in terms of Standard Deveiation
    # Table 4.1 Column D
    df['Spikes'] = (df.Close - df.Close.shift(1)) / df.StdDevLog.shift(1)

    # Calculate Annualized Volatility from Standard Deviation of Log Change over sliding Window
    # Annual Volatility
    # Number of trading periods in a year.
    # For daily pricing data, trading days in a year
    # For 1 minute pricing data, trading minutes in a year
    df['AnnVol'] = df.StdDevLog * sqrt(yearly_trading_period)
    
    # Only calculate intraday and overnight price spikes and volatility for daily data
    if (yearly_trading_period == 252):
        df = calc_price_spikes_annual_vol_intraday(df, window, yearly_trading_period)
        df = calc_price_spikes_annual_vol_overnight(df, window, yearly_trading_period)
    
    # Calculate Volume Weighted Average Price
    #df['vwap'] = (((df['Open'] + \
    #                df['High'] + \
    #                df['Low'] + \
    #                df['Close']) / 4.0) * \
    #              df['Volume']\
    #             ).rolling(window=window, \
    #                       min_periods=1).sum() / df['Volume'].rolling(window=window, \
    #                                                                   min_periods=1).sum()
    
    df['vwap'] = (((df['Open'] + df['High'] + df['Low'] + \
                    df['Close']) / 4.0) * df['Volume']).cumsum() / df['Volume'].cumsum()

    # Intraday Reversal (in percentage)
    # (High - Low) / Close
    df['intraday_reversal'] = 100.0 * (df['High'] - df['Low']) / df['Close']
    df['intraday_reversal_MA'] = df.intraday_reversal.rolling(window).mean()

    # Change Index back to original column
    df = df.set_index(index_name)
    
    return df

def calc_price_spikes_annual_vol_intraday(df, window=20, yearly_trading_period=252):
    # ------------------------------------------------------------------------
    # Calculate Intraday Price Spikes and Annual Volatility Open to Close (O-C)
    # ------------------------------------------------------------------------
    # Calculate Intraday Volatility (OC)

    # Calculate Percent Change from Open to Close on the same day
    df['PctChng_intraday'] = ((df.Close / df.Open) - 1.0) * 100.0

    # Calculate Log Change from Open to Close
    df['LogChng_intraday'] = np.log(df.Close / df.Open)

    # Standard Deviation of LogChng over a specified time period (sliding window)
    df['StdDevLog_intraday'] = df.LogChng_intraday.rolling(window).std() * df.Close

    # Price Spikes in terms of Standard Deveiation
    df['Spikes_intraday'] = (df.Close - df.Open) / df.StdDevLog_intraday.shift(1)

    # Calculate Annualized Volatility from Standard Deviation of Log Change over sliding Window
    # Annual Volatility
    # Number of trading periods in a year.
    # For Intraday, Number of Trading Days in a Year x Number of hours in a day / Trading hours in a day
    intraday_trading_period = yearly_trading_period * 24 / 6.5
    df['AnnVol_intraday'] = df.StdDevLog_intraday * sqrt(intraday_trading_period)
    
    return df

def calc_price_spikes_annual_vol_overnight(df, window=20, yearly_trading_period=252):
    # ------------------------------------------------------------------------
    # Calculate Overnight Price Spikes and Annual Volatility Close to Open (C-O)
    # ------------------------------------------------------------------------
    # Calculate Overnight Volatility (C-O)

    # Calculate Percent Change from Close to Open on the next day
    df['PctChng_overnight'] = ((df.Open / df.Close.shift(1)) - 1.0) * 100.0

    # Calculate Log Change from Close Open on the next day
    df['LogChng_overnight'] = np.log(df.Open / df.Close.shift(1))

    # Standard Deviation of LogChng over a specified time period (sliding window)
    df['StdDevLog_overnight'] = df.LogChng_overnight.rolling(window).std() * df.Open

    # Price Spikes in terms of Standard Deveiation
    df['Spikes_overnight'] = (df.Open - df.Close.shift(1)) / df.StdDevLog_overnight.shift(1)

    # Calculate Annualized Volatility from Standard Deviation of Log Change over sliding Window
    # Annual Volatility
    # Number of trading periods in a year.
    # [TODO]: Need to account for holidays and weekends in non-trading hours
    # For Overnight, Number of Trading Days in a Year x Number of hours in a day / Non-Trading hours in a day
    overnight_non_trading_period = yearly_trading_period * 24 / 17.5
    df['AnnVol_overnight'] = df.StdDevLog_overnight * sqrt(overnight_non_trading_period)

    return df

# Calculate Intraday Volume adjusted prices (VAP)
def calc_vap(df):
    # Intraday price adjusted for volume
    # df = intraday stock data downloaded from Yahoo! Finance
    # Returns:
    #.    df with calculated 'vap' column
    vol = df.Volume.values
    opn = df.Open.values
    cls = df.Close.values
    hi = df.High.values
    lo = df.Low.values
    
    vap = (((opn + hi + lo + cls) / 4.0) * vol).cumsum() / vol.cumsum()
    
    return df.assign(vap = vap)

# Detect Support and Resistance
# Fractal candlestick pattern
# Bullish Fractal Pattern
def is_support(df, i):
    cond1 = df.Low[i] < df.Low[i-1]
    cond2 = df.Low[i] < df.Low[i+1]
    cond3 = df.Low[i+1] < df.Low[i+2]
    cond4 = df.Low[i-1] < df.Low[i-2]
    
    return cond1 and cond2 and cond3 and cond4

# Bearish Fractal Pattern
def is_resistance(df, i):
    cond1 = df.High[i] > df.High[i-1]
    cond2 = df.High[i] > df.High[i+1]
    cond3 = df.High[i+1] > df.High[i+2]
    cond4 = df.High[i-1] > df.High[i-2]
    
    return cond1 and cond2 and cond3 and cond4

# Check if support ot resistance level exist
def is_far_from_level(value, levels, df):
    ave = np.mean(df.High - df.Low)
    
    return np.sum([abs(value - level) < ave for _, level in levels]) == 0

# Detect Breakout
# Returns true when previous candle opens below a level and the last candle opens and closes below that level
def has_breakout(levels, previous, last):
    for _, level in levels:
        cond1 = previous.Open < level
        cond2 = (last.Open > level) and (last.Low > level)
        
    return cond1 and cond2

# Get Dividend history for a security
def get_dividend(security):
    # Get dividend from downloaded corporate action CSV file
    try:
        data_type = 'actions'
        # get latest data file for specified security
        file_name = max(glob.iglob('data/{}_{}.csv'.format(security, data_type)), key=os.path.getctime)
        actions = pd.read_csv(file_name, parse_dates=True, infer_datetime_format=True)

        dividend = actions[actions.action == 'DIVIDEND'].head(1).value.values[0]
        
        # Get dividend directly from Yahoo
        #dividends = web.DataReader(security, 'yahoo-dividends')
        #dividend = dividends.iloc[0]['value']
    except:
        dividend = 0.0
        
    return dividend
    
# Get 3 months T-bill rate from FRED
def get_risk_free_rate():
    # Get 3month T-bill rate from FRED for Risk Free Rate
    try:
        tb3ms = web.DataReader('TB3MS', 'fred')
        risk_free_rate = tb3ms.iloc[-1].values[0] / 100.0
    except:
        risk_free_rate = 0.01

    return risk_free_rate
```

```python
##### Functions - Option #####

# Get ATM Option Data for all expirations
def get_atm_options(options_df):
    # Function to find atm rows for each group of expiry
    def get_atm_rows(df):
        trading_period = 252
        # Row in each group closest to ATM strike
        idx = df['Strike'].sub(df['Underlying_Price']).abs().idxmin()
        # ATM strike for each group
        atm_strike = df.loc[idx]['Strike']
        x = df[df.Strike == atm_strike]
        # Calculate 1 Standard Deviaiton 1 Day Price Change
        x['PriceChange1SD'] = x.Underlying_Price * x.IV / sqrt(trading_period)
        x['PriceChange1SDdte'] = x.PriceChange1SD * np.sqrt(abs(x.dte))

        return x

    # Group option pricing df by expiartion date and apply get_atm_rows function
    # return the resulting dataframe
    return options_df.groupby(options_df.Expiry, sort=True).apply(get_atm_rows).reset_index(drop=True)

# Nearest Expiration Option Data
def get_nearest_expiration_options(options_df):
    # Find nearest expiration date, ATM strike, Stock Price, and ATM Implied Volatility
    # options_df = options data downloaded from Yahoo! Finance using yfinance
    # Returns:
    #.    nearest_expiry = nearest option expiration date as date
    #.    atm_strike = ATM Strike for nearest expiry
    #.    stock_price = Stock Price
    #.    atm_iv = ATM Option Implied Volatility
    #.    nearest_expiry_df = Dataframe with all options date for the nearest expiration
        
    # Find the nearest expiration date
    nearest_expiry = options_df.Expiry.min()
    nearest_expiry_df = options_df[options_df.Expiry == nearest_expiry]

    # Find the strikes nearest to underlying price (ATM/near-ATM)
    # Find row with lowest absolute difference between Strike and Underlying Price
    idx = nearest_expiry_df['Strike'].sub(nearest_expiry_df['Underlying_Price']).abs().idxmin()

    # Nearest strike for nearest expiration date
    atm_strike = nearest_expiry_df.loc[idx]['Strike']
    stock_price = nearest_expiry_df.loc[idx]['Underlying_Price']

    return nearest_expiry, atm_strike, stock_price, nearest_expiry_df

# Calculate option value and greeks using  Black-Scholes-Morten Model
def calc_bsm_stats(df, dividend=0.0, risk_free_rate=0.01):
    # Input
    #. A dataframe of option data with single row
    #.. df.Expiry
    #.. df.Underlying_Price
    #.. df.Strike
    #.. df.IV
    #.. df.Bid, df.Ask for implied volatility calc
    #.. df.Type
    #
    # Output
    #. Return dataframe with added columns for BSM stats
    #.. bs_price
    #.. bs_delta
    #.. bs_gamma
    #.. bs_vega
    #.. bs_theta
    #.. bs_rho
    #.. bs_elasticity
    #.. bs_iv
        
    # option data
    d = datetime.strptime(df.Expiry,'%Y-%m-%d')
    maturity_date = ql.Date(d.day, d.month, d.year)

    spot_price = df.Underlying_Price
    strike_price = df.Strike
    volatility = df.IV
    
    dividend_rate = dividend / df.Underlying_Price

    if (df.Type == 'call'):
        option_type = ql.Option.Call
    else:
        option_type = ql.Option.Put
    
    # Option evaluation date
    calculate_date = ql.Date_todaysDate()

    # Collect quoted market data
    u = ql.SimpleQuote(spot_price)
    r = ql.SimpleQuote(risk_free_rate)
    sigma = ql.SimpleQuote(volatility)

    # Get Black-Scholes-Merton model
    day_count = ql.Actual365Fixed()
    calendar = ql.UnitedStates()

    ql.Settings.instance().evaluationDate = calculate_date

    # Construct the European Option
    payoff = ql.PlainVanillaPayoff(option_type, strike_price)
    exercise = ql.EuropeanExercise(maturity_date)
    european_option = ql.VanillaOption(payoff, exercise)

    # Black-Scholes-Merton
    riskFreeCurve = ql.FlatForward(0, calendar, ql.QuoteHandle(r), day_count)
    volatility = ql.BlackConstantVol(0, calendar, ql.QuoteHandle(sigma), day_count)
    dividendCurve = ql.FlatForward(0, calendar, dividend_rate, day_count, ql.Quarterly)

    flat_ts = ql.YieldTermStructureHandle(riskFreeCurve)
    dividend_yield = ql.YieldTermStructureHandle(dividendCurve)
    flat_vol_ts = ql.BlackVolTermStructureHandle(volatility)

    bsm_process = ql.BlackScholesMertonProcess(ql.QuoteHandle(u), 
                                               dividend_yield, 
                                               flat_ts, 
                                               flat_vol_ts)

    engine = ql.AnalyticEuropeanEngine(bsm_process)
    european_option.setPricingEngine(engine)
    
    # Compute theoretical price and greeks
    df['bs_price'] = european_option.NPV()
    df['bs_delta'] = european_option.delta()
    df['bs_gamma'] = european_option.gamma()
    df['bs_vega'] = european_option.vega()
    df['bs_theta'] = european_option.thetaPerDay()
    df['bs_rho'] = european_option.rho()
    df['bs_elasticity'] = european_option.elasticity()
    
    # Calculate Implied Volatility if Option Bid and Ask are available
    try:
        # Calculate mid price from Bid Ask quotes
        option_bid_ask_mid = (df.Bid + df.Ask) / 2.0
        df['bs_iv'] = european_option.impliedVolatility(option_bid_ask_mid, bsm_process)
    except:
        df['bs_iv'] = None

    return df

# Calculate Option Stats as Underlying Price change
def get_option_stats_with_price(df, underlying_range, dividend=0.0, risk_free_rate=0.01):

    # For a single option, calculate option values and greeks as underlying price changes, 
    # using Black-Scholes_Merton Model

    # Get the BSM option model

    # option data
    d = datetime.strptime(df.Expiry.values[0],'%Y-%m-%d')
    maturity_date = ql.Date(d.day, d.month, d.year)

    spot_price = df.Underlying_Price.values[0]
    strike_price = df.Strike.values[0]
    imp_vol = df.IV.values[0]
    dividend_rate = dividend / spot_price

    if (df.Type == 'call').bool():

        option_type = ql.Option.Call
    else:
        option_type = ql.Option.Put

    # Option evaluation date
    calculate_date = ql.Date_todaysDate()

    # Collect quoted market data
    u = ql.SimpleQuote(spot_price)
    r = ql.SimpleQuote(risk_free_rate)
    sigma = ql.SimpleQuote(imp_vol)

    # Get Black-Scholes-Morten Model
    day_count = ql.Actual365Fixed()
    calendar = ql.UnitedStates()

    ql.Settings.instance().evaluationDate = calculate_date

    # Construct the European Option
    payoff = ql.PlainVanillaPayoff(option_type, strike_price)
    exercise = ql.EuropeanExercise(maturity_date)
    european_option = ql.VanillaOption(payoff, exercise)

    # Black-Scholes-Merton
    riskFreeCurve = ql.FlatForward(calculate_date, ql.QuoteHandle(r), day_count)
    volatility = ql.BlackConstantVol(calculate_date, calendar, ql.QuoteHandle(sigma), day_count)
    dividendCurve = ql.FlatForward(calculate_date, dividend_rate, day_count, ql.Quarterly)

    flat_ts = ql.YieldTermStructureHandle(riskFreeCurve)
    dividend_yield = ql.YieldTermStructureHandle(dividendCurve)
    flat_vol_ts = ql.BlackVolTermStructureHandle(volatility)

    bsm_process = ql.BlackScholesMertonProcess(ql.QuoteHandle(u), 
                                               dividend_yield, 
                                               flat_ts, 
                                               flat_vol_ts)

    engine = ql.AnalyticEuropeanEngine(bsm_process)
    european_option.setPricingEngine(engine)

    result_df = pd.DataFrame()
    
    for spot_price in underlying_range:
        u.setValue(spot_price)
        # add to result_df
        result_df = pd.concat([result_df, 
                               pd.DataFrame({'Strike': strike_price,
                                  'Expiry': df.Expiry,
                                  'Type': df.Type,
                                  'Symbol': df.Symbol, 
                                  'stock_price': spot_price,
                                  'imp_vol': imp_vol,
                                  'dte': df.dte,
                                  'bs_price': european_option.NPV(),
                                  'bs_delta': european_option.delta(),
                                  'bs_gamma': european_option.gamma(),
                                  'bs_vega': european_option.vega(),
                                  'bs_theta': european_option.thetaPerDay(),
                                  'bs_rho': european_option.rho(),
                                  'bs_elasticity': european_option.elasticity()
                                 })
                              ], ignore_index=True)

    # Reset Spot Price to current stock price
    u.setValue(df.Underlying_Price.values[0])
    
    return result_df

# Calculate Option Stats as Implied Volatility Changes
def get_option_stats_with_volatility(df, volatility_range, dividend=0.0, risk_free_rate=0.01):

    # For a single option, calculate option values and greeks as volatility changes, 
    # using Black-Scholes_Merton Model

    # Get the BSM option model

    # option data
    d = datetime.strptime(df.Expiry.values[0],'%Y-%m-%d')
    maturity_date = ql.Date(d.day, d.month, d.year)

    spot_price = df.Underlying_Price.values[0]
    strike_price = df.Strike.values[0]
    imp_vol = df.IV.values[0]
    dividend_rate = dividend / spot_price

    if (df.Type == 'call').bool():

        option_type = ql.Option.Call
    else:
        option_type = ql.Option.Put

    # Option evaluation date
    calculate_date = ql.Date_todaysDate()

    # Collect quoted market data
    u = ql.SimpleQuote(spot_price)
    r = ql.SimpleQuote(risk_free_rate)
    sigma = ql.SimpleQuote(imp_vol)

    # Get Black-Scholes-Morten Model
    day_count = ql.Actual365Fixed()
    calendar = ql.UnitedStates()

    ql.Settings.instance().evaluationDate = calculate_date

    # Construct the European Option
    payoff = ql.PlainVanillaPayoff(option_type, strike_price)
    exercise = ql.EuropeanExercise(maturity_date)
    european_option = ql.VanillaOption(payoff, exercise)

    # Black-Scholes-Merton
    riskFreeCurve = ql.FlatForward(0, calendar, ql.QuoteHandle(r), day_count)
    volatility = ql.BlackConstantVol(0, calendar, ql.QuoteHandle(sigma), day_count)
    dividendCurve = ql.FlatForward(0, calendar, dividend_rate, day_count, ql.Quarterly)

    flat_ts = ql.YieldTermStructureHandle(riskFreeCurve)
    dividend_yield = ql.YieldTermStructureHandle(dividendCurve)
    flat_vol_ts = ql.BlackVolTermStructureHandle(volatility)

    bsm_process = ql.BlackScholesMertonProcess(ql.QuoteHandle(u), 
                                               dividend_yield, 
                                               flat_ts, 
                                               flat_vol_ts)

    engine = ql.AnalyticEuropeanEngine(bsm_process)
    european_option.setPricingEngine(engine)

    result_df = pd.DataFrame()

    for imp_vol in volatility_range:
        sigma.setValue(imp_vol)
        # add to result_df
        result_df = pd.concat([result_df, 
                               pd.DataFrame({'Strike': strike_price,
                                  'Expiry': df.Expiry,
                                  'Type': df.Type,
                                  'Symbol': df.Symbol, 
                                  'stock_price': spot_price,
                                  'imp_vol': imp_vol,
                                  'dte': df.dte,
                                  'bs_price': european_option.NPV(),
                                  'bs_delta': european_option.delta(),
                                  'bs_gamma': european_option.gamma(),
                                  'bs_vega': european_option.vega(),
                                  'bs_theta': european_option.thetaPerDay(),
                                  'bs_rho': european_option.rho(),
                                  'bs_elasticity': european_option.elasticity()
                                 })
                              ], ignore_index=True)

    # Reset Implied Volatility to current volatility
    sigma.setValue(df.IV.values[0])
    
    return result_df

# Calculate Option Stats each day until expiration
def get_option_stats_with_dte(df, dividend=0.0, risk_free_rate=0.01):

    # For a single option, calculate option values and greeks each day until expiration, 
    # using Black-Scholes_Merton Model

    # Get the BSM option model

    # option data
    d = datetime.strptime(df.Expiry.values[0],'%Y-%m-%d')
    maturity_date = ql.Date(d.day, d.month, d.year)

    spot_price = df.Underlying_Price.values[0]
    strike_price = df.Strike.values[0]
    imp_vol = df.IV.values[0]
    dividend_rate = dividend / spot_price

    if (df.Type == 'call').bool():

        option_type = ql.Option.Call
    else:
        option_type = ql.Option.Put

    # Option evaluation date
    calculate_date = ql.Date_todaysDate()

    # Collect quoted market data
    u = ql.SimpleQuote(spot_price)
    r = ql.SimpleQuote(risk_free_rate)
    sigma = ql.SimpleQuote(imp_vol)

    # Get Black-Scholes-Morten Model
    day_count = ql.Actual365Fixed()
    calendar = ql.UnitedStates()

    # might need this line
    # The result on expiration is given by setting the following (the option expires in-the-money):
    ql.Settings.instance().includeReferenceDateEvents = True
    
    
    ql.Settings.instance().evaluationDate = calculate_date

    # Construct the European Option
    payoff = ql.PlainVanillaPayoff(option_type, strike_price)
    exercise = ql.EuropeanExercise(maturity_date)
    european_option = ql.VanillaOption(payoff, exercise)

    # Black-Scholes-Merton
    # changed to 0, calendar from calculate_date
    riskFreeCurve = ql.FlatForward(0, calendar, ql.QuoteHandle(r), day_count)
    volatility = ql.BlackConstantVol(0, calendar, ql.QuoteHandle(sigma), day_count)
    dividendCurve = ql.FlatForward(0, calendar, dividend_rate, day_count, ql.Quarterly)

    flat_ts = ql.YieldTermStructureHandle(riskFreeCurve)
    dividend_yield = ql.YieldTermStructureHandle(dividendCurve)
    flat_vol_ts = ql.BlackVolTermStructureHandle(volatility)

    bsm_process = ql.BlackScholesMertonProcess(ql.QuoteHandle(u), 
                                               dividend_yield, 
                                               flat_ts, 
                                               flat_vol_ts)

    engine = ql.AnalyticEuropeanEngine(bsm_process)
    european_option.setPricingEngine(engine)

    result_df = pd.DataFrame()
    
    # number of days between today and expiration
    tot_days = maturity_date - calculate_date
    
    for day in calendar.businessDayList(calculate_date, maturity_date):
        ql.Settings.instance().evaluationDate = day
        # add to result_df
        result_df = pd.concat([result_df, 
                               pd.DataFrame({'Strike': strike_price,
                                  'Expiry': df.Expiry,
                                  'Type': df.Type,
                                  'Symbol': df.Symbol, 
                                  'stock_price': spot_price,
                                  'imp_vol': imp_vol,
                                  'dte': maturity_date - day,
                                  'bs_price': european_option.NPV(),
                                  'bs_delta': european_option.delta(),
                                  'bs_gamma': european_option.gamma(),
                                  'bs_vega': european_option.vega(),
                                  'bs_theta': european_option.thetaPerDay(),
                                  'bs_rho': european_option.rho(),
                                  'bs_elasticity': european_option.elasticity()
                                 })
                              ], ignore_index=True)
    
    return result_df
```

```python
# Option Summary Statistics
def get_option_stats(df, x='Open_Int'):
    total = df[x].sum()
    call = df[df.Type == 'call'][x].sum()
    put = df[df.Type == 'put'][x].sum()
    if min(call, put) > 0:
        cp_ratio = '{:,.2f}:{:,.2f}'.format(call/min(call, put), put/min(call, put))
    else:
        cp_ratio = None
        
    return total, cp_ratio

rows01 = ['Open Interest', 'Recent Trade']
rows02 = ['Total', 'C:P Ratio']
columns = ['All', 'Nearest Expiry', 'ATM', 'ATM Nearest Expiry']

index = pd.MultiIndex.from_product([rows01, rows02], names=['Universe', 'Stats'])
option_stats_df = pd.DataFrame(index=index, columns=columns)

option_stats_df.loc['Open Interest', 'All'] = get_option_stats(options_df, x='Open_Int')
option_stats_df.loc['Open Interest', 'ATM'] = get_option_stats(atm_df, x='Open_Int')
option_stats_df.loc['Open Interest', 'Nearest Expiry'] = get_option_stats(nearest_expiry_df, x='Open_Int')
option_stats_df.loc['Open Interest', 'ATM Nearest Expiry'] = get_option_stats(nearest_atm_df, x='Open_Int')
option_stats_df.loc['Recent Trade', 'All'] = get_option_stats(options_df, x='Vol')
option_stats_df.loc['Recent Trade', 'ATM'] = get_option_stats(atm_df, x='Vol')
option_stats_df.loc['Recent Trade', 'Nearest Expiry'] = get_option_stats(nearest_expiry_df, x='Vol')
option_stats_df.loc['Recent Trade', 'ATM Nearest Expiry'] = get_option_stats(nearest_atm_df, x='Vol')

# Recent Trade / Open Interest ratio may indicate unusual activity
option_stats_df.loc[('Unusual Activity', 'Vol/OI'), ] = 100 * option_stats_df.loc[('Recent Trade', 'Total'), ] / option_stats_df.loc[('Open Interest', 'Total'), ]

print('{} Option Summary Data'.format(security))
display(option_stats_df.T.style.format(formatter={('Open Interest', 'Total'):'{:,.0f}', ('Recent Trade', 'Total'):'{:,.0f}', ('Unusual Activity', 'Vol/OI'):'{:,.2f}%'}))
```

```python
# Consecutive Up periods and consecutive Down periods
price_df = price_df.assign(Up = price_df.PctChng >= 0,
                                  Down = price_df.PctChng < 0)

price_df = price_df.assign(
    Up_cons = price_df.Up * (
        price_df.Up.groupby(
            (price_df.Up != price_df.Up.shift()).cumsum()).cumcount() + 1), 
    Down_cons = price_df.Down * (
        price_df.Down.groupby(
            (price_df.Down != price_df.Down.shift()).cumsum()).cumcount() + 1))

freq_df = pd.DataFrame(price_df.Up_cons.value_counts(normalize=False))
freq_df = freq_df.assign(Up_cons_freq = price_df.Up_cons.value_counts(normalize=True))
freq_df = freq_df.assign(Down_cons = price_df.Down_cons.value_counts(normalize=False))
freq_df = freq_df.assign(Down_cons_freq = price_df.Down_cons.value_counts(normalize=True))

print('Frequency of consecutive UP and DOWN days for {}'.format(security))
Up_days = price_df.loc[last_trade_date, 'Up_cons']
Dn_days = price_df.loc[last_trade_date, 'Down_cons']

if (Up_days > 0):
    nextd_up_prob = 100 * freq_df.loc[Up_days].Up_cons / freq_df.loc[Up_days - 1].Up_cons
    print('{} Consecutive {:,.0f} Up days.\nNext day UP Probability: {:,.2f}%'.format(last_trade_date, Up_days, nextd_up_prob))
else:
    nextd_dn_prob = 100 * freq_df.loc[Dn_days].Down_cons / freq_df.loc[Dn_days - 1].Down_cons
    print('{} Consecutive {:,.0f} Down days.\nNext day DOWN Probability: {:,.2f}%'.format(last_trade_date, Dn_days, nextd_dn_prob))

display(freq_df \
        .style
        .set_properties(**{'color':'black', 'background-color':'greenyellow'}, subset=pd.IndexSlice[max(price_df.loc[last_trade_date, 'Up_cons'], price_df.loc[last_trade_date, 'Down_cons'])])
        .format(formatter=({'Up_cons': '{:,.0f}', 'Down_cons': '{:,.0f}'}), precision=4)
       )
```

## Special Events

### Key Concepts

* Option pricing based on assumption that market trades continuously. In reality, market is open for 6.5 hours and closed for 17.5 hours on normal trading days, and closed for 65.5 hours each weekend. The relative magnitude of these distortions increases sharply as expiration approaches.

* Market efficiently responds to these distortions by depressing and inflating option prices in predictable ways.

* News events often create brief inefficiencies that represent outstanding trading opportunities that sometimes persist for several hours.

### Introduction

1. **Implied volatility skew** or smile is designed to guard against outsized downward spikes that are not comprehended by traditional pricing thoery. It causes OTM Puts to be aggressively priced with significantly higher IV than ATM options.

2. **Operation Hours**. The final weekend of each expiration represents 33% of the time remaining before contracts expire. More significantly, it represents 39% of time left before market close at 4:00pm on Friday and the contracts can no lomger be traded. The market responds by depressing option prices near close on Friday and reinflating on Monday morning. IV falls near close on Friday to compensate for upcoming weekend time decay and rises Monday morning to restore prices to normal level. This anomaly becomes even more dramatic Thursday before expiration when overnight time frame consumes 17.5 hours of remaining 24 hours before options stop trading.

3. **Planned high-impact events** such as earning announcement, FOMC meeting and interest rate announcement, presidential speech, OPEC meeting, etc. represent another distorting force that confounds traditional option pricing models. IV often rises sharply before earnings to accomodate the perceived risk of a surprise. After the announcement, regardless of results, IV falls rapidly as stock settles at an appropriate new price. There is no precise way of knowing whether the IV before announcement is too low, correct, or too high. Furthermore, Put:Call parity inflates both sides while market tends to perceive risk as being on either upside or downside. It causes one the two sides to be dramatically overpriced regardless of the magnitude of surprise. After price settle, IV falls and new trading opportunities emerge based on the new trend.

### Unavoidable Volatility Distortions

The Long Weekend before expiration causes market to be closed for 89.5 hours representing 47% of the 192 hours left before expiration Friday. Also, the day before Long Weekend tend to be unusually calm as traders leave early. Option buyers are hesitant to overpay, guaranteed to suffer enormous time decay before next trading opportunity after Long Weekend. We can verify, using an option calculator that \\$100 Call trading with 50% IV on a \\$100 stock will collapse from \\$3.19 to \\$2.47 during final long weekend. Prices tend to collapse near the close of the final trading day. Sellers normally offset their risk with structured hedges such as Long OTM Puts and short positions. Investors who typically purchase options as market closes on last trading day before long weekend are not penalized for 3 day gap because they paid dramatically reduced prices. For example, trading a 1:2 ratio consisting of 1 long \\$60 call and 2 short \\$65 calls for Research in Motion over Easter weekend 2009, returned more than 80% in just 6.5 hours without virtually no change in underlying stock price. The trade cost \\$0.89 at the open and sold for \\$1.61 at the close.

Options are superior trading instruments over stocks for day trading because they can generate profit from volatility collapse or time decay without any change in the underlying price. Amomalies of this sort are rare but their enormous profit potential cannot be ignored. 

### Event Trading

Carefully chosen market-moving events present the same sort of opportunity. Although most are not as profitable as the expiration week example, many occur on regular basis. For example, the OIH etf which represents a basket of oil services and exploration companies is among the most sensitive to the weekly petroleum status report. 

**The end of the downtrend is typically marked by two upward sloping parallel lines in the spike chart, that indicate successively larger upward spikes and correspondingly lower downward spikes.**

In many cases the market immediately reacts to a specific item in the report before reversing direction after all the information is analyzed. These reversals can be much more severe than the original trend ignited by the news.

Reversals can be also be conmfusing when they are caused by short covering. "Short squeezes" are very common in event driven trading situations. On the option side these scenarios normally result in widening bid-ask spreads. It is wise to check the short interest data before trading the results of a major news event.

### Summary

* A pricing distortion related to a long weekend just prior to expiration week.
* Capitalize on the market's adjustment to a relatively large amount of new information.

### Basic Concepts

Many investors prefer bear markets because the declines are often much steeper than the increases associated with a traditional bull market. Markets can crash down but they rarely crash up. Reducing market exposure by trading in very brief time frames can help during bear markets. Blindly hanging on with a bullish or bearish view is a flawed strategy.

Option trading in turbulent times can also be diffuclt. IV rise sharpely, making simpple long put or call positions unreasonably expensive, and the risks associated with naked short positions are simply too large for any conservative investor. Investors can avoid all these pitfalls by entering market at very specific times and structuring trades that capitalize on well-characterized pricing anomalies and distortions. Option traders can use these distortions to structure positions that are both statistically advantaged and direction neutral. For example, long straddles that have potential to generate profit from a large price change in either direction. Options day trading focuses only on the underlying mathematics.

Short term trading strategies can be simple. It is often possible to simply **trade the news**. A short-term trader who anticipated major events, such as geo-political, would have a large impact on the market could have purchased an <u>ATM straddle</u> on index (DIA, SPY, QQQ). Options provide a distinct advantage because they allow the construction of direction neutral positions. Option traders can decide that a position is underpriced because it doesn't represent the risk of an upcoming news event. A structured position, on the basis of this view, can profit from a large swing in either direction.

Option prices tend to shrink near the close on Friday as market discounts weekend time decay into the price, i.e. IV tends to decrease on Friday afternoon.

Option day traders can also profit from **implied volatility swings**. An example is the long weekend before the last week of expiration. The long weekend represents 3 of the remaining 7 days in expiration cycle. Option buyers are hesitant to over pay for contracts that are about to lose a significant value while market was closed. The result is persistent falling IV from the open to the close. This market efficiency represents a trading opportunity because it condenses 3 days of time decays near the end of an expiration cycle into a single trading day. A <u>ratio spread</u>, for example 10 long $60 calls and 20 short $65 calls, opened in the morning and closing in afternoon can give significant return even if stock doesm't move much.

#### Recent Changes in the Option Market Support Day Trading

Near the end of a brief but significant rally, buyers often become less aggressive and lower their bid prices, causing the bid-ask spread to widen. This change is often a better indicator that a rally is ending. Day traders who watch the Level II quotes often capitalize on this opportunity to sell options that quickly lose value if stock reverses direction and corrects slightly downward.

* <u>Long straddles</u> should be initiated only with stock trading close to a strike price where put and call deltas are equal.Day traders must often wait for the right conditions to avoid placing trades with a directional character. Closer strike price spacing helps solve this problem.

#### Unstable Markets Provide Unique Advantages to Option Traders

Excessive market turmoil changes the dyanamics of option pricing:

1. Higher IV across all option contracts.

2. Steepened IV skew from low to high strike prices.

3. Large IV swings, i.e. IV has become more volatile.

The turmoil is apparent from charts of the CBOE Volatility Index (VIX). Higher VIX, for example 80, can completely destablize option prices during the rapid market collapse. Such changes represent new opportunities for structuring trades encompassing a range of strike prices and expirations. In percentage terms, the IV skew between ATM and OTM options are typically flat before and after crash. However, in absolute value terms, IV increases much more dramatically after crash. This distortion creates a much steeper curve.

* The steeper IV curve envrionment allows complex positions to be structured across many strikes and increases liquidity, for example a <u>Ratio Trade</u>.

##### General Procedure for Calculating One Standard Deviation Price Change

$P = $ Current Stock Price

$D = $ Number of Trading Periods in a Year = 12 months = 52 weeks = 252 days = 181,440 2-minute intervals

$IV = $ Annual Implied Volatility, in fraction (not as percentage)

$\Delta P_{std} = $ One Standard Deviation Daily Price Change

$$ \Delta P_{std} = P * \frac{IV}{\sqrt{D}} = P * \frac{IV}{\sqrt{252}} \approx P * \frac{IV}{16}$$

$\Delta P_{Tstd} = $ One Standard Deviation Price Change over a Fixed Period $T$

$$ \Delta P_{Tstd} = P * \frac{IV}{\sqrt{\frac{D}{T}}} = P * IV * {\sqrt{\frac{T}{D}}} = P * \frac{IV}{\sqrt{D}} * {\sqrt{{T}}} =  \Delta P_{std} * {\sqrt{{T}}}$$

### Trading Volatility Distortions

#### Key Concepts

* Investors often build complex 3D maps that relate IV to calendar information and price of the underlying security. Distortions and anomalies in these maps represent trading opportunities in the form of mispriced volatility.

* Overnight (close-to-open), intraday (open-to-close), and traditional (close-to-close) measures of volatility can vary dramatically. These differences can be used to identify mispriced options and structure statistically advantaged trades.

* Volatility distortions sometimes vary by weekday. These calendar effcets can help time entry and exit points for certain types of short-term trades.

#### The Implied Volatility Surface

Volatility swings are the norm and not the exception. It is important to estimate historical volatility so that trading price of an option can be compared to its theoretical value.

When an option is heavily traded, it can be assumed that market forces will push the price close to the correct theoretical value. Surprisingly, this price is often very different from the value obtained by calculating the recent historical volatility of stock.

Investors often build complex 3D maps that relate IV to expiration calendar and underlying price. These maps, or volatility surfaces are an important tool for determining the fair value of an option. A 3D map can be constructed with strike prices on the x-axis, term to expiration on the y-axis, and IV on the z-axis (vertical axis). The map would be specific to a particular date and set of market conditions. At a later date when market and the individual security display similar charactersitics, the map can be retrieved for comparison to actual trading prices. These maps often reveal repeating oscillations - stochastics - that can be used to predict rising and falling IV.

Option traders who use the Black-Scholes formula in their option pricing models must continuously adjust their volatility assumptions to arive at actual trading price of an option.

The IV skew or "smile" was designed to accommodate the risk of a rapid price collapse, causes lower strikes to be priced higher with higher IV. As a result, OTM puts/calls are relatively more expensive that ITM puts/calls.

An investor who expects the market to fall might, therefore, purchase VIX calls or sell VIX puts.

IV surfaces can be especially valuable to select the best combination of options for a calendar spread because they offer insight into differences between various expiration/strike combo.

The VIX is unusual because it represents the IV priced into options on S&P500 stocks. Volatility of VIX, therefore, is best thought of as the volatility of volatility. Stocks and indexes display IV profiles with much flatter skews and term structure than VIX.

Because IV priced into options can differ significantly from actual volatility of underlying, it is important to have access to both. IV varies between strikes and expirations; historical volatility varies with the length of the window used for its calculations. Volatility can also vary significantly from one time of the day to another.

Volatility from close to open (overnight) tends to be smaller than from open to close (intraday). Certain days are also more volatile - most notably, days that include scheduled announcements such as earnings.

#### Calculating Fair Volatility

Prices rise when buyers are more aggressive. They fall when sellers are more aggressive. When both parties disagree and neither side feels pressured to execute a trade, bid-ask spreads widen. These dynamics alone determine the price of a financial instrument.

The principal goal of an option trader is to arbitrage subtle differences between implied and fair volatility by structuring positions that capitalize on those differences. The underlying must rise or fall more or less than the amount priced into some position of a trade to generate a profit.

No single volatility value can be used to explain option prices across different strikes and expirations. Unfortunately, option prices are very sensitive to volatility.

We can define volatility as a 1 standard deviation (StdDev) price change, in percent, at the end of 1 year. Using the normal distribution, we can assume a 68% chance that the final price will fall within +/- 1 StdDev. Some adjustment for interest rates is also necessary, then a 1 StdDev price change at year end will be $ (1 + r) * Price * StdDev $.

Another key characteristic of volatility is that it is proportional to the sqaure root of time. This dynamic allows us to extrapolate short term volatility from annual volatility. For example, because there are 252 trading days in a year, we can estimate a \\$100 stock with 40% annual volatility will exhibit daily volatility of 0.40/SQRT(252) = 2.5%. This value simplifies to \\$2.50 (2.5% the stock price), also the value of a 1 StdDev daily price change.

Historical volatility is calculated as the standard deviation of the short-term returns multiplied by the square root of the number of trading days in a year, also known as annualization factor. This is a reasonable approach when the goal is to structure trades that will remain in place for days or weeks. However, when trading in shorter time frames such as a single day, close-to-close changes are too granular because they merge together intraday, overnight, and weekend information.

For example, one stock might exhibit frequent large opening price spikes followed by relatively intraday stability, whereas the other stock might tend to open unchanged before becoming volatile. The first reacts to overnight events around the world; the second reacts to news events that occur during the trading day. *The first stock might be a candidate for overnight long straddles initiated near the closing bell; the second might be a candidate for same trade placed after market stabilizes around 10am in the morning.*

Most significant are the two intervals that make up a trading day: the intraday interval when market opens at 9:30 and  closes at 16:00 (O-C), and overnight time frame from the close at 16:00 to the open at 9:30 (C-O). The intraday interval is 6.5 hours, and the overnight interval is 17.5 hours.

Differential volatility has significant implications because it creates trading opportunities. For most stocks, intraday price changes are dramatically more volatile than overnight changes. #### Trading Volatility Distortions

The most sensible approach is to structure positions early in the day and hold them until they are profitable or the market closes. The goal is to avoid owning dramatically overpriced options during the long overnight time frame when time decay dominates and volatility is diminished. Option contracts experience constant time decay that accelerates as expiration approaches.

These dynamics might be interpreted as a reason to sell options and profit from time decay. Unfortunately such strategies require being short options through the turbulent intraday time frame when actual volatility can be several times larger than IV. Worse still, occasional large overnight price spikes can easily erase accumulated profits when market is closed and trade can't be closed.

*Strategies that involve opening and closing short positions at strategic times - before and after weekends, for example, are flawed because bid-ask spreads often erase most of the profit. The market also responds to impending time decay with downward adjustments to option prices. The effcet can be seen as declining IV and widening bid-ask spreads near the close on Friday.*

<u>With few exceptions, selling volatility and time decay is a risky business. One of the most significant exception is expiration week, when time decay accelerates and each overnight interval represents major portion of the remaining time.</u>

<u>Another problem with selling time decay is that profits can only be earned slowly and market exposure is continuous. Conversely, long positions designed to profit from a large price change can generate profit quickly but can only lose money very slowly.</u>

## Strategy: Delta-neutral Long Straddle and Strangles

Either position, straddle or strangle, should be structured so that the overall trade is delta neutral. A delta-neutral position can always be structured at any stock price by adjusting the number of puts and calls. Generally speaking, strangles tend to be a better choice than straddles because the trades are less expensive and the percent gain will always be larger for any size underlying price change.

## Strategy 2: Alternate Trade Structures - Backspreads

Other trade structures can also be used to leverage underpriced volatility. Optimally, the trade should be capable of generating substantial amounts of profit in a short time frame with minimal downside risk.

Long straddles fit the profile because they can generate large amounts of profit from a 2 StdDev price change with only the risk of modest time decay in the event that the stock price remains stable.

**Ratio Backspreads (Reverse Ratios)** composed of short options at a near strike and a large quantity of long options at a far strike represent another excellent alternative. An investor who <u>anticpiates a large upward price change</u> for a $97 stock might, for example, create a position constsing of 10 short \\$100 calls and 20 long \\$105 calls. <u>If the position is delta-neutral, small price change will have little effect. However, a sharp rise in stock price will generate a profit as the delta of the far strike grows faster than that of the near strike.</u> It is the inverse of the more common **Forward Ratio**. Forward Ratios are designed to generate profit from slow movement and time decay. Forward call ratios that are long the near strike and short the far strike **fail** under following circumstances:

* The stock rises beyond the break-even point by expiration.
* The stock rises a modest amount early in the life cycle of the trade.
* Market conditions cause IV to rise for all strikes.
* Positive news causes a shift in the shape of the IV skew, and the far strike becomes relatively more expensive.
* The stock falls sharply and both sides of trade become worthless, excess premium paid is lost.
* Most forward ratios are debit trades where the short side only partially covers the cost of the long side.
* The stock falls sharply but recovers. IV rises across all strikes and remains relatively high after recovery causing far strike to become disproportionately expensive, very common during periods of market instability.

<u>Backspreads (Reverse Ratios) are profitable under each of the above circumstances. They generate profit if the stock rises too far or too quickly, or if IV rises.</u>

Forward Ratios can only make money slowly but lose it quickly; Reverse Ratios can make money quickly but only lose it slowly. Although originally intended as a ST trade, the lack of time decay risk allows Reverse Ratio position to be kept open, without penalty for an extended length of time. Asymmetry is the trade-off as significant profit can be generated in only one direction if the trade is held until expiration. Maximum profit in the other direction is limited to the initial credit of the trade. It is always beneficial to close such trades, keep the profit, and start again. Because the long side is twice as large as the short side, it experiences slightly more time decay. If both strikes remain OTM and trade collapses to zero, the initial credit will be realized as profit.

These dynamics create a very low-risk situation and add a level of flexibility that long straddles do not have. Strangles behave differently because they suffer from constant time decay.

In summary, short time frames and modest price changes form an excellent environment for long strangles. If, however, the stock price remains unchanged, a strangle will lose due to time decay. Conversely, backspreads generate smaller profits from modest price changes but are more resilient. A backspread can generate a reasonable profit with the stock price unchanged.

Despite the huge price changes, excessively high IV can render long straddles nearly worthless. Backspreads are a more logical choice for this trading environment. In the environment with overall <u>average intraday reversal, measured as daily (High - Low)/Close above 10\%</u>, all structured positions are difficult to manage, which emphasizes the importance of promptly closing profitable trades. Whenever possible, investors close enough of the profitable part of a trade to pay for initial cost.

> McMillan, Lawrence G., Options As A Strategic Investment, 4th Ed., Penguin Putnam, 2002. Chapter 13: Reverse Spread

This strategy presents a reasonable method of attempting to capitalize on a large stock movement with little tie-up of collateral. Generally, the strategist would seek out **volatile stocks** for implementation of this strategy, because he would want as much potential movement as possible by the time the calls expire.

An increase in implied volatility will be beneficial to a backspread strategy, while a decrease in implied volatility will be slightly harmful to the spread. However, since risk is limited in a backspread, such a decrease in implied volatility would not have catastrophic conse- quences unless one had overcommitted his funds to one position. (pp. 781)

> Natenberg, Sheldon, Option Volatility and Pricing, 2nd Ed., McGraw Hill, 2015. Chapter 11: Volatility Spread. pp. 179.

A ratio spread where more options are purchased than sold is sometimes referred to as backspread. Regardless of whether the spread consists of calls or puts, this type of spread always wants movement in the underlying market (+gamma, theta) and/or an increase in implied volatility (+vega).

> Cohen, Guy, Bible of Options Strategies : The Definitive Guide for Practical Trading Strategies, 2nd Ed., FT Press, 2016. Chapter 6: Levergaed Strategies.

Description: The call ratio backspread is an exciting strategy that enables you to make accelerated profits provided that the **stock moves sharply upwards**. Increasing volatility is very helpful because youre net long in calls. **The worst thing that can happen is that the stock doesnt move at all**. Even a sharp move down can be profitable, or at the very least preferable to no movement at all.

Step In: Try to ensure that **the stock is trending upward, and is above a clearly identifiable area of support.**

Step Out: In any event, look to unravel the trade at least one month **before expiration**, either to capture your profit or to contain your losses.

Outlook: With call ratio backspreads, your outlook is aggressively bullishyou are looking for increasing volatility with the stock price moving explosively upwards.

Rationale: To execute a bullish trade for little to no cost while reducing your maximum risk. You are looking for the stock to rise significantly.

Appropriate Time Period to Trade: You will be safer to choose a longer time to expiration, preferably at least six months.

Risk Profile:

* Maximum Risk: [Difference in Strikes] - [Net Credit Received]
* Maximum Reward: [Uncapped]
* Breakeven Down: [Lower Strike + Net Credit]
* Breakeven Up: [Higher Strike + (Difference in Strikes x Number of Short Calls)/(Number of Long Calls - Number of Short Calls) - [Net Credit Received]]

If the stock price remains around the higher Strike, we make our maximum loss. If it moves explosively upwards, we make uncapped profit.

Greeks:

* Delta: Delta (speed) is at its greatest the more stock rises. The more stock falls, slower the position becomes, and hence delta slows to zero.
* Gamma: Gamma (acceleration) peaks around higher Strike, illustrating the position's major turning point and the fastest rate of change.
* Theta: Time decay is predominantly harmful to the position unless stock falls dramatically. You want as much time to be right as possible because you are looking for such a large move.
* Vega: Volatility is predominantly helpful to the position.
* Rho: Higher interest rates are generally helpful to the position, particularly when stock rises.

Exiting the Trade/ Mitigating a Loss: By selling one of the bought calls, you will be creating a bear call spread, a less aggressive trade but one that also involves a change of directional sentiment.

```python
# Call Backspreads (Reverse Ratio)

# Initialization and constants
backspread_ratio = 2.0

def ratio_backspread(target_df, Type='call', stock_price=stock_price, backspread_ratio=backspread_ratio, dividend=dividend, risk_free_rate=risk_free_rate):
    # Initialization and constanst
    trading_period = 252 # number of trading days in a year
    pos_val_multiplier = 100 # multiplier for dollar value of position
    save_flag = True # Used for Call spread only
    
    result_dict = dict() # result initialization
    
    # Reduce target_df to one of the Option type
    df = target_df[target_df.Type == Type].reset_index()
    
    # Strikes range to calculate ratio backspreads
    ratio_strikes = sorted(df.Strike.unique())
    
    # 1.5+ StdDev move in stock price
    
    # Find strikes bracketing stock price
    atm_target = min(ratio_strikes, key=lambda x: abs(x - stock_price))    
    strike_lower = atm_target if (atm_target <= stock_price) else ratio_strikes[ratio_strikes.index(atm_target)-1]
    strike_higher = atm_target if (atm_target >= stock_price) else ratio_strikes[ratio_strikes.index(atm_target)+1]
    
    # Calculate IV at stock price
    # Stock IV will be slightly different from ATM IV if ATM Strike is not same as stock price

    if (strike_lower == strike_higher):
        stock_iv = df[(df.Strike == atm_target)].IV.values[0]
    else:
        stock_iv = (df[(df.Strike == strike_lower)].IV.mul(stock_price - strike_lower).values[0] + df[(df.Type == Type) & (df.Strike == strike_lower)].IV.mul(stock_price - strike_lower).values[0]) / (strike_higher - strike_lower)

    # Stock price if stock moves 1.5 StdDev Up or Down
    stock_price_down_15sd = stock_price - (1.5 * stock_price * stock_iv / np.sqrt(trading_period))
    stock_price_up_15sd = stock_price + (1.5 * stock_price * stock_iv / np.sqrt(trading_period))
        
    # Create a List of Options Delta
    deltas = sorted(df.bs_delta.dropna().unique())

    # For each option delta find short and long strike
    for long_delta in deltas:
        # Short strike can't have absolute delta greater than 1
        if (abs(long_delta) <= 1.0/backspread_ratio):
            short_delta = min(deltas, key=lambda x: abs(x - (backspread_ratio * long_delta)))
            
            short_strike = df[df.bs_delta == short_delta].Strike.values[0]
            long_strike = df[df.bs_delta == long_delta].Strike.values[0]
            
            # For each Long-Short pair, change underlying price and estimate profitability
            long_df = df[df.Strike == long_strike]
            short_df = df[df.Strike == short_strike]
            
            # underlying price range
            # For table only need 3 points
            underlying_range = np.sort(np.asarray([stock_price_down_15sd, stock_price, stock_price_up_15sd]))

            long_price_df = get_option_stats_with_price(df=long_df, underlying_range=underlying_range, dividend=dividend, risk_free_rate=risk_free_rate)
            short_price_df = get_option_stats_with_price(df=short_df, underlying_range=underlying_range, dividend=dividend, risk_free_rate=risk_free_rate)
            
            long_price_df = long_price_df.assign(bs_price = long_price_df.bs_price.mul(backspread_ratio),
                                     bs_delta = long_price_df.bs_delta.mul(backspread_ratio),
                                     bs_gamma = long_price_df.bs_gamma.mul(backspread_ratio),
                                     bs_vega = long_price_df.bs_vega.mul(backspread_ratio),
                                     bs_theta = long_price_df.bs_theta.mul(backspread_ratio),
                                     bs_elasticity = long_price_df.bs_elasticity.mul(backspread_ratio))

            short_price_df = short_price_df.assign(bs_price = short_price_df.bs_price.mul(-1),
                                       bs_delta = short_price_df.bs_delta.mul(-1),
                                       bs_gamma = short_price_df.bs_gamma.mul(-1),
                                       bs_vega = short_price_df.bs_vega.mul(-1),
                                       bs_theta = short_price_df.bs_theta.mul(-1),
                                       bs_elasticity = short_price_df.bs_elasticity.mul(-1))
        
            # Combine Long and Short price data
            ratio_df = pd.concat([long_price_df, short_price_df], ignore_index=True)

            # Group by Stock Price to get combined value of option price and greeks
            grouped_price = ratio_df.groupby('stock_price', sort=True, as_index=False).agg(np.sum)
        
            # Save the plotting data if short strike is at ATM just below ATM
            # Note: Plot data is continuously overwritten until short strikes at ATM
            cond1 = (Type == 'call') and (short_strike <= stock_price) and save_flag # first entry is correct
            cond2 = (Type == 'put') and (short_strike >= stock_price) # last entry is correct
            
            if (cond1 or cond2):
                plot_long_df = long_df.copy()
                plot_short_df = short_df.copy()
                                
                # For Calls, break after first entry
                if (cond1):
                    save_flag = False

            # position value with 1.5 SD stock price change            
            ratio_val_0sd = grouped_price.loc[grouped_price.stock_price == stock_price].bs_price.values[0]
            ratio_val_15sd_down = grouped_price.loc[grouped_price.stock_price == stock_price_down_15sd].bs_price.values[0]
            ratio_val_15sd_up = grouped_price.loc[grouped_price.stock_price == stock_price_up_15sd].bs_price.values[0]

            # P&L
            ratio_val_gain_down = ratio_val_15sd_down - ratio_val_0sd
            ratio_val_gain_up = ratio_val_15sd_up - ratio_val_0sd

            
            # Final Output
            result_dict[len(result_dict)] = {'S_strike': short_strike,
                                             'L_strike': long_strike,
                                             'S_delta': short_delta,
                                             'L_delta': long_delta,
                                             'net_delta': (backspread_ratio * long_delta) - short_delta,
                                             'cost': ratio_val_0sd * pos_val_multiplier,
                                             'val_15sd_up': ratio_val_15sd_up * pos_val_multiplier,
                                             'val_15sd_dn': ratio_val_15sd_down * pos_val_multiplier,
                                             'pnl_up': ratio_val_gain_up * pos_val_multiplier,
                                             'pnl_dn': ratio_val_gain_down * pos_val_multiplier,
                                            }

    # convert result dict to df
    result_df = pd.DataFrame.from_dict(result_dict, orient='index')
    
    print('{} {} Delta-neutral 1:{:,.0f} {} Ratio Backspread Profitability'.format(security, df.Expiry.unique()[0], backspread_ratio, Type))
    print('Stock Price: ${:,.2f} ATM {} Strike: {:,.2f} Value Multiplier: {:,.0f}'.format(stock_price, Type, atm_target, pos_val_multiplier))
    print('+/-1.5 SD Stock Price Change. Down: ${:,.2f}, Up: ${:,.2f}'.format(stock_price_down_15sd, stock_price_up_15sd))
    #return result_df, plot_short_price_df, plot_long_price_df, plot_grouped_price
    return result_df, plot_short_df, plot_long_df, stock_price_down_15sd, stock_price_up_15sd
```
